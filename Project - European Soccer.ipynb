```python
# Let's join match, league, countries and team table and see which are the home team, and away teams and their respective goals,
# on each matches.
detailed_matches = pd.read_sql("""SELECT Match.id, 
                                        Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season, 
                                        stage, 
                                        date,
                                        HT.team_long_name home_team,
                                        AT.team_long_name away_team,
                                        home_team_goal,
                                        away_team_goal                                   
                                    FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                JOIN Team HT on HT.team_api_id = Match.home_team_api_id
                                JOIN Team AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name = 'Spain'
                                ORDER by date
                                LIMIT 10;""", conn)
detailed_matches
```

```python
# Let's create few addtional features from the given data and see it. 
leages_by_season = pd.read_sql("""SELECT Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season,
                                        count(distinct stage) AS number_of_stages,
                                        count(distinct HT.team_long_name) AS number_of_teams,
                                        avg(home_team_goal) AS avg_home_team_scors, 
                                        avg(away_team_goal) AS avg_away_team_goals, 
                                        avg(home_team_goal-away_team_goal) AS avg_goal_dif, 
                                        avg(home_team_goal+away_team_goal) AS avg_goals, 
                                        sum(home_team_goal+away_team_goal) AS total_goals                                       
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name in ('Spain', 'Germany', 'France', 'Italy', 'England')
                                GROUP BY Country.name, League.name, season
                                HAVING count(distinct stage) > 10
                                ORDER BY Country.name, League.name, season DESC
                                ;""", conn)
leages_by_season 
# Ref: Kaggle/dimarudov
```

```python
# Dividing the matches in 3(Under 500, Under 1000 & more than 1000) categories based on total goals scored by the teame
leages_by_season_with_goals_cat = pd.read_sql("""SELECT Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season,
                                        count(distinct stage) AS number_of_stages,
                                        count(distinct HT.team_long_name) AS number_of_teams,
                                        avg(home_team_goal) AS avg_home_team_scors, 
                                        avg(away_team_goal) AS avg_away_team_goals, 
                                        avg(home_team_goal-away_team_goal) AS avg_goal_dif, 
                                        avg(home_team_goal+away_team_goal) AS avg_goals, 
                                        sum(home_team_goal+away_team_goal) AS total_goals,   
                                        
                                CASE   WHEN sum(home_team_goal+away_team_goal) < 500 THEN 'Under 500 goals'
                                       WHEN sum(home_team_goal+away_team_goal) BETWEEN 500 and 1000 THEN 'Under 1000 goals'
                                    ELSE 'More than 1000 goals'
                                    END AS total_goals_category
                                        
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name in ('Spain', 'Germany', 'France', 'Italy', 'England')
                                GROUP BY Country.name, League.name, season
                                HAVING count(distinct stage) > 10
                                ORDER BY Country.name, League.name, season DESC
                                ;""", conn)
leages_by_season_with_goals_cat
```

```python
# Let's see all the mathces with more than 1000 goals
More_than_total_1k_goals = pd.read_sql("""SELECT Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season,
                                        count(distinct stage) AS number_of_stages,
                                        count(distinct HT.team_long_name) AS number_of_teams,
                                        sum(home_team_goal+away_team_goal) AS total_goals                                       
                                
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                GROUP BY Country.name, League.name, season
                                HAVING count(distinct stage) > 10 and total_goals>=1000
                                ORDER BY total_goals DESC
                                ;""", conn)
More_than_total_1k_goals 
```

```python
# Let's join both Player and Player_attribues table and see which are the top ten players based on overall ratings
top_players = pd.read_sql("""SELECT P.player_name, PA.overall_rating, PA.potential, 
                            P.height, P.weight,
                            PA.preferred_foot, PA.attacking_work_rate,
                            PA.defensive_work_rate
                            
                            FROM Player as P
                            INNER JOIN Player_Attributes as PA on P.id = PA.player_api_id
                            
                            ORDER BY PA.overall_rating desc
                            LIMIT 10""", conn)
top_players
```

```python
# As we want distinct names hence, we'll use group by clause
top_players_groupby_name = pd.read_sql("""SELECT DISTINCT P.player_name, MAX(PA.overall_rating) as max_rating, PA.potential, 
                            P.height, P.weight,
                            PA.preferred_foot, PA.attacking_work_rate,
                            PA.defensive_work_rate
                            
                            FROM Player as P
                            INNER JOIN Player_Attributes as PA on P.id = PA.player_api_id
                            
                            GROUP BY P.player_name
                            
                            ORDER BY PA.overall_rating desc
                            LIMIT 10""", conn)
top_players_groupby_name
```

```python
# Just for fun, let's see if Ronaldo and Messi are present in the data and if what are their stats
lets_find_ronaldo_and_messi = pd.read_sql("""SELECT DISTINCT P.player_name, MAX(PA.overall_rating) as max_rating, PA.potential, 
                            P.height, P.weight,
                            PA.preferred_foot, PA.attacking_work_rate,
                            PA.defensive_work_rate
                            
                            FROM Player as P
                            INNER JOIN Player_Attributes as PA on P.id = PA.player_api_id
                             
                            WHERE P.player_name LIKE '%nald%' OR P.player_name LIKE '%messi%'
                            
                            ORDER BY PA.overall_rating desc
                            LIMIT 10""", conn)
lets_find_ronaldo_and_messi
```
